```python
import pandas as pd
import numpy as np
```

```python
from scholarly import scholarly

# Retrieve the author's data, fill-in, and print
# Get an iterator for the author results
search_query = scholarly.search_author('Steven A Cholewiak')
# Retrieve the first result from the iterator
first_author_result = next(search_query)
scholarly.pprint(first_author_result)

# Retrieve all the details for the author
author = scholarly.fill(first_author_result )
scholarly.pprint(author)

# Take a closer look at the first publication
first_publication = author['publications'][0]
first_publication_filled = scholarly.fill(first_publication)
scholarly.pprint(first_publication_filled)

# Print the titles of the author's publications
publication_titles = [pub['bib']['title'] for pub in author['publications']]
print(publication_titles)

# Which papers cited that publication?
citations = [citation['bib']['title'] for citation in scholarly.citedby(first_publication_filled)]
print(citations)
```

```python
papers = [a for a in scholarly.search_pubs('NLP')]
```

```python
#papers[0]['pub_url']
```

```python
#master_list = []

for pub in papers:
    #print(pub)
    title = pub['bib']['title']
    author = pub['bib']['author']
    pub_year = pub['bib']['pub_year']
    venue = pub['bib']['venue']
    abstract = pub['bib']['abstract']
    try:
        pub_url = pub['pub_url']
    except:
        pub_url = None
        print('No pub url')
    author_id = pub['author_id']
    num_citations = pub['num_citations']
    gsrank = pub['gsrank']
    url_scholarbib = pub['url_scholarbib']
    
    temp_dict = {'title' : title,
                'author' : author,
                'pub_year' : pub_year,
                'venue' : venue,
                'abstract' : abstract,
                'pub_url' : pub_url,
                'author_id' : author_id,
                'num_citations' : num_citations,
                'gsrank' : gsrank,
                'url_scholarbib' : url_scholarbib}
    master_list.append(temp_dict)
```

```python
NLP_df = pd.DataFrame(master_list)
```

```python
NLP_df.drop_duplicates().shape
```

```python
NLP_df.drop_duplicates(inplace = True)
```

```python
NLP_df.to_csv('NLP_paper_info.csv')
```
