## PDF parsing to get references

```python
import os
import sys
import re

import PyPDF2

import pandas as pd
```

```python
df = pd.read_csv('../data/arxiv_papers_full.csv', index_col='Unnamed: 0')
df
```

```python
def pdf_to_text(path):
    with open(path, 'rb') as file:
        reader = PyPDF2.PdfReader(file, strict=False)
        pdf_text = []
        
        for page in reader.pages:
            content = page.extract_text()
            pdf_text.append(content)
    return pdf_text
```

```python
fname = os.listdir('../data/PDFs') #fname : List contain pdf documents names.
print(fname)
#fname: must be sorted.
fname.sort(key=lambda f: (re.sub('\D', '', f)))
length = len(fname)
```

```python
df['filename'] = df['PDF URL'].apply(lambda x: x.split('/')[-1])
fname = df.filename.to_list()
fname
```

```python
full_text = []
for i in range(len(fname)): #Repeat each operation for each document.
    text_output = pdf_to_text(('../data/PDFs/{}.pdf').format(fname[i])) #Extract text with PDF_to_text Function call
    full_text.append(text_output)
```

```python
df['Full Text'] = full_text
```
