```python
import pandas as pd
import numpy as np
```

```python
from scholarly import scholarly
```

```python
# Get an iterator for the author results
#search_query = scholarly.search_keyword('natural language processing')
```

```python
papers = [a for a in scholarly.search_pubs('NLP')]
```

```python
papers#[0]['pub_url']
```

```python
#master_list = []

for pub in papers:
    #print(pub)
    title = pub['bib']['title']
    author = pub['bib']['author']
    pub_year = pub['bib']['pub_year']
    venue = pub['bib']['venue']
    abstract = pub['bib']['abstract']
    try:
        pub_url = pub['pub_url']
    except:
        pub_url = None
        print('No pub url')
    author_id = pub['author_id']
    num_citations = pub['num_citations']
    gsrank = pub['gsrank']
    url_scholarbib = pub['url_scholarbib']
    
    temp_dict = {'title' : title,
                'author' : author,
                'pub_year' : pub_year,
                'venue' : venue,
                'abstract' : abstract,
                'pub_url' : pub_url,
                'author_id' : author_id,
                'num_citations' : num_citations,
                'gsrank' : gsrank,
                'url_scholarbib' : url_scholarbib}
    master_list.append(temp_dict)
```

```python
NLP_df = pd.DataFrame(master_list)
```
